```python
import mysql.connector
import os


# Source .env file
from dotenv import load_dotenv
load_dotenv()

passwordENV = os.environ.get('LOCAL_DB_PASSWORD')

# Connection configuration
configLocal = {
    'user': 'dan',
    'password': passwordENV,
    'host': 'localhost',
    'port': '3307',
    'database': 'main'
}
# Connect to MariaDB
connectionLocal = mysql.connector.connect(**configLocal)
cursor = connectionLocal.cursor()
```

```python
# TODO DC: filter out non-english sources

qry = "SELECT id, datetime, newsource, headline, pagetitle, pageurl FROM main.headlines LIMIT 10000"

data = []
cursor.execute(qry)
for (id, datetime, newssource, headline, pagetitle, pageurl) in cursor:
  # print("{}, {}, {}, {}, {}, {}".format(
  #   id, datetime, newssource, headline, pagetitle, pageurl
  # ))
  
  # [str, str, datetime.datetime]
  data.append([
    str(headline),
    str(newssource),
    datetime,
  ])
  
cursor.close()
```
